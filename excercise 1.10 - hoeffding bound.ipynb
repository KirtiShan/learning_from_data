```python
from random import random

def flip_coin():
    return 0 if random() < 0.5 else 1
```

```python
import numpy as np

coins = np.empty([1000, 10])
v = np.empty([100000, 3])
```

```python
for m in range(100000):
    for i in range(1000):
        for k in range(10):
            coins[i, k] = flip_coin()

    means = np.mean(coins, axis=1)

    v[m,0] = means[0]
    v[m,1] = means[int(1000*random())]
    v[m,2] = min(means)
```

```python
import matplotlib.pyplot as plt

plt.hist(v, bins=11)
```

```python
est = np.empty([11, 3])
e = np.empty(11)

for i in range(11):
    e[i] = i/10
    est[i] = np.mean((abs(v-0.5) >= e[i]), axis=0)
```

```python
plt.plot(e, est)
```
